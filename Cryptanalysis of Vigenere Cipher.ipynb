```python
import matplotlib.pyplot as plt

def split_bins(bin_text):
    frequency = {}
    total_characters = len(bin_text)
    for char in bin_text:
        if char in frequency:
            frequency[char] += 1
        else:
            frequency[char] = 1

    for char, count in frequency.items():
        frequency[char] = (count / total_characters) * 100
    
    sorted_frequency = dict(sorted(frequency.items(), key=lambda item: item[1], reverse=True))
    
    return sorted_frequency

def plot(frequency):
    characters = list(frequency.keys())
    occurrences = list(frequency.values())
    
    plt.bar(characters, occurrences)
    plt.xlabel('Character')
    plt.ylabel('Occurrences')
    plt.title('Frequency Analysis')
    plt.show()

def frequency_analysis(ciphertext):
    bins = {'1': '', '2': '', '3': '', '0': ''}
    filtered_text = ''.join([char for char in ciphertext if 32 <= ord(char) <= 126])
    for i, char in enumerate(filtered_text):
        modulo_value = i % 4
        if str(modulo_value) not in bins:
            bins[str(modulo_value)] = ''
        bins[str(modulo_value)] += char
    
    for key, value in bins.items():
        print(f'Bin {key}: {value}')
        frequency = split_bins(value)
        plot(frequency)

    
ct="""NnyAagx9[r!Acy4#yh}4^&$(yu%'h&,'nru0^y@Am{w*yg(Ajxu+loyAix45nk%2_&!#ek(My}y6ym'#my!#hj41l&"#ly|'m&,+nn4&_t('yl'+hm}0a&+'ak)#no$0&&u0^&*5og!.s&z'_j(A\4&[hv.ct{A`u'Ajru0n&z1ij49cz|Abku&yy*$gk')_jBANny;yiu0yg!5i&x+pk47hjy4qg)'l&z1l&z1ij@Agu''yv'1`ow+_t).s&)*[t41nny4yju$\r}0a&x7]q(Myg#&ysu;yg!5i&(6_g!A`u$&yl'1g&x+po#)yh}4^y45oi|A[y4%iu)"""
frequency_analysis(ct)
```
